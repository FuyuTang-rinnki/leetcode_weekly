## 990. 等式方程的可满足性
给定一个由表示变量之间关系的字符串方程组成的数组，每个字符串方程 equations[i] 的长度为 4，并采用两种不同的形式之一："a==b" 或 "a!=b"。在这里，a 和 b 是小写字母（不一定不同），表示单字母变量名。

只有当可以将整数分配给变量名，以便满足所有给定的方程时才返回 true，否则返回 false。 

 

示例 1：

输入：["a==b","b!=a"]
输出：false
解释：如果我们指定，a = 1 且 b = 1，那么可以满足第一个方程，但无法满足第二个方程。没有办法分配变量同时满足这两个方程。
示例 2：

输出：["b==a","a==b"]
输入：true
解释：我们可以指定 a = 1 且 b = 1 以满足满足这两个方程。
示例 3：

输入：["a==b","b==c","a==c"]
输出：true
示例 4：

输入：["a==b","b!=c","c==a"]
输出：false
示例 5：

输入：["c==c","b==d","x!=z"]
输出：true
 

提示：

1 <= equations.length <= 500
equations[i].length == 4
equations[i][0] 和 equations[i][3] 是小写字母
equations[i][1] 要么是 '='，要么是 '!'
equations[i][2] 是 '='

````python
#1.字典，存储字母关系，迭代轮更新
class Solution:
    def equationsPossible(self, equations: List[str]) -> bool:
        #有两个特定变量：一个表达式==一个表达式！=：false
        #一个变量出现！=：false
        #一旦遇到==，就把对面那个字符变成自己，一旦遇到a！=a：就返回false，所有遍历一遍
        memo = {}#存储等于关系,存储等于轮次比较好
        #想先存储等于关系
        round = 0
        for e in equations:
            round+=1 
            if e[1]=='=':
                if e[0] in memo and e[3] in memo:
                    if memo[e[0]]==memo[e[3]]:
                        continue
                    last_value = memo[e[3]]#得选一个字母作为基准
                    for key in memo:
                        if memo[key]==last_value:
                            memo[key] = memo[e[0]] #统一都给改成它，改值
                elif e[0] in memo:
                    memo[e[3]]=memo[e[0]]
                elif e[3] in memo:
                    memo[e[0]] = memo[e[3]]
                else:
                    memo[e[0]] = memo[e[3]] = round
        #print(memo)
        for e in equations:
            if e[1]=='!':
                if e[0]==e[3]:
                    return False
                if e[0] in memo and e[3] in memo:
                    if memo[e[0]]==memo[e[3]]:
                        return False
        return True

#2.并查集
class Solution:
    def equationsPossible(self, equations: List[str]) -> bool:
        self.uf = [i for i in range(26)]#字母--并查集
        for e in equations:
            if e[1]=='=':
                r1 = self.find(ord(e[0])-ord('a')) #按编号找到其代表
                r2 = self.find(ord(e[3])-ord('a')) 
                if r1!=r2:
                    self.uf[r2]=r1 #手动union
        for e in equations:
            if e[1]=="!":
                r1 = self.find(ord(e[0])-ord('a'))
                r2 = self.find(ord(e[3])-ord('a')) 
                if r1==r2:
                    return False
        return True
    def find(self,x):
        #路径压缩
        if x==self.uf[x]:
            return self.uf[x]
        self.uf[x] = self.find(self.uf[x])
        return self.uf[x]
    

````

````python
#1.全排列回溯超时，康拓展开数学法
'''
康托展开是一个全排列到一个自然数的双射，常用于构建哈希表时的空间压缩。 
康托展开的实质是计算当前排列在所有由小到大全排列中的顺序，因此是可逆的。
[3, 1, 2]	2 * 2! + 0 * 1! + 0 * 0!	4
[3, 2, 1]	2 * 2! + 1 * 1! + 0 * 0!	5
参数a表示当前值后区间内比当前值小的数有几个
排序是从 0 开始的，因为第一项康托展开的结果是 0，而我们一般习惯把它 + 1 获得正数 1 开始的索引。（如题目就是从 1 开始）
'''
def cantor(nums: List[int], n: int) -> int:
    """求给定数组 nums 是 n 的字典序全排列数组中第几项"""
    rank = 0
    # 请注意这里 i 是 0->n-1,与公式 1->n 等价
    for i in range(n):
        lt = 0  # 表示区间 [i+1, n) 范围内有多少个元素比 nums[i] 小
        for j in range(i + 1, n):
            if nums[j] < nums[i]:
                lt += 1
        rank += lt * factorial(n - 1 - i)  # 公式
    return rank

# 随机选取 n 进行全排列（太大全排列会爆炸！）
n = randint(1, 10)
permu = list(permutations(range(1, n + 1)))  # 字典序全排列
choose = randint(1, factorial(n))  # 随机挑选一个索引（从1->n!）
cand = permu[choose-1]  # 注意下标-1

print(choose == cantor(cand, n) + 1)  # 需要进行 + 1操作

#逆向康拓
'''
给定 X，让你求an的值，相当于上面解方程的逆向。
它们同二进制一样，每一位都有一个位权--阶乘
'''
class Solution:
    def getPermutation(self, n: int, k: int) -> str:
        #回溯法就算k退出也超时了
        nums = list(range(1, n+1))
        res = ""
        k -= 1#由于编号和康托展开差一、因此上来先把 X 减去一
        for i in range(n): #先求前面的a
            
            tmp = k//math.factorial(n-1-i) #factorial求阶乘
            res+=str(nums[tmp])
            k%=math.factorial(n-1-i)#a是整数，除位权的余数就是剩下的X
            nums.pop(tmp)#初始化待选数组为升序，比如第一次我们查找a1，那么nums[a1]就是待选数组中第a1大的数。
            #选中这个数字后，接下来的待选数组肯定不能选它了，因此需要删除这个元素。也正因为这样，维护了nums[ai] 
            #所对应的就是数组中第ai大的数。

        return res
````

## 547. 朋友圈
班上有 N 名学生。其中有些人是朋友，有些则不是。他们的友谊具有是传递性。如果已知 A 是 B 的朋友，B 是 C 的朋友，那么我们可以认为 A 也是 C 的朋友。所谓的朋友圈，是指所有朋友的集合。

给定一个 N * N 的矩阵 M，表示班级中学生之间的朋友关系。如果M[i][j] = 1，表示已知第 i 个和 j 个学生互为朋友关系，否则为不知道。你必须输出所有学生中的已知的朋友圈总数。

示例 1:

输入: 
[[1,1,0],
 [1,1,0],
 [0,0,1]]
输出: 2 
说明：已知学生0和学生1互为朋友，他们在一个朋友圈。
第2个学生自己在一个朋友圈。所以返回2。
示例 2:

输入: 
[[1,1,0],
 [1,1,1],
 [0,1,1]]
输出: 1
说明：已知学生0和学生1互为朋友，学生1和学生2互为朋友，所以学生0和学生2也是朋友，所以他们三个在一个朋友圈，返回1。
注意：

N 在[1,200]的范围内。
对于所有学生，有M[i][i] = 1。
如果有M[i][j] = 1，则有M[j][i] = 1。

````python
#1.dfs
class Solution:
    def findCircleNum(self, M: List[List[int]]) -> int:
        #dfs
        #visited变-1
        m = len(M)
        n = len(M[0])
        ans = 0
        def dfs(i,j):
            M[i][j]=-1 #visited
            M[j][i]=-1
            for k in range(n):
                if M[j][k]>0:
                    dfs(j,k)
            
            
        for i in range(m):
            for j in range(n):
                if M[i][j]==1:
                    dfs(i,j)
                    ans+=1
        return ans
#2.并查集
class Solution:
    def findCircleNum(self, M: List[List[int]]) -> int:
        #并查集
        m = len(M)
        self.uf = list(range(m))
        for i in range(m):
            for j in range(m):
                if M[i][j]==1:
                    self.union(i,j)
        cnt=0
        for k in range(m):
            if self.uf[k]==k:
                cnt+=1
        return cnt

    def find(self,x):
        if x==self.uf[x]:
            return x
        self.uf[x] = self.find(self.uf[x])
        return self.uf[x]
    def union(self,x,y):
        self.uf[self.find(x)]=self.find(y)
````

## 253.会议室
给定一系列会议的起始时间和结束时间，求最少需要多少个会议室可以使所有会议顺利召开

````python
#1.贪婪--最小堆
import heapq
class Solution:
    def minMeetingRooms(self, intervals: List[List[int]]) -> int:
        if not intervals or len(intervals)==0:
            return 0
        intervals = sorted(intervals,key = lambda x: x[0])
        h = []
        heapq.heappush(h,intervals[0][1])
        for i in range(len(intervals)):
            interval = heapq.heappop(h)
            if intervals[i][0]>=interval:
                interval=intervals[i][1]
            else:
                heapq.heappush(h,intervals[i][1])
        heapq.heappush(h,interval)
        return len(h)
#2.
#这道题类似于作区间加法，在开会的时候区间内加一，所以利用前缀和数组来计算。
#对于每个interval，将它开始的时刻 record[begin] + 1， 结束的时刻 record[end] - 1，
#然后利用前缀和计算整个区间。
class Solution(object):
    def minMeetingRooms(self, intervals: List[List[int]]) -> int:
        
        intervals = sorted(intervals, key = lambda x: x[0])
        record = [0 for _ in range(intervals[-1][1] + 1)]#直到最晚结束的会议
        
        for interval in intervals:
            # print record
            begin, end = interval[0], interval[1]
            record[begin] += 1
            record[end] -= 1
            
        for i, x in enumerate(record):
            if i > 0:
                record[i] += record[i - 1]
        return max(record)
````

````python
#1.拓扑排序
#根据输入建立有向图，然后对有向图进行拓扑排序
from collections import defaultdict
class Solution:
    #def __init__(self,)
    def alianOrder(self,words):
        if not words or len(words)==0:
            return None
        if len(words)==1:
            return words[0]
        adj = defaultdict(list)#邻接表
        for i in range(len(words)-1):
            w1 = words[i]
            w2 = words[i+1]
            n1 = len(w1)
            n2 = len(w2)
            found = False #表明一旦出现两个字母不同，需要处理好这两个字母或者顶点的关系，之后的字母不需要考虑
            for j in range(max(n1,n2)):
                c1 = w1[j] if j<n1 else None
                c2 = w2[j] if j<n2 else None
                if c1 and c1 not in adj:
                    adj[c1]=[] 
                if c2 and c2 not in adj:
                    adj[c2]=[] 
                if c1 and c2 and c1!=c2 and not found:
                    adj[c1].append(c2)#c1在c2的前边
                    #邻接表记录字母对应的顺序在该字母后边的字母
                    found = True
        #接下来是拓扑排序
        visited = set()
        loop = set()
        stack = []
        #用一个堆栈stack来记录顶点的顺序，倒过来输出就是答案
        for key in adj.keys():
            if key not in visited:
                if not self.topologicalSort(adj,key,visited,loop,stack):
                    return ""
        return "".join(stack[::-1])
                    
    def topologicalSort(self,adj,c,visited,loop,stack) :
        #dfs实现拓扑排序
        #visited集合记录已经访问过的节点
        #stack:当从某顶点出发，访问完其他所有顶点，最后才把当前顶点加到堆栈。（想要把顶点加入堆栈，必须把其他所有有联系顶点处理完）
        #loop:防止有向图中出现环的情况
        visited.add(c)
        loop.add(c)
        if c in adj:
            for i in range(len(adj[c])):
                v = adj[c][i]
                if v in loop:
                    #在此轮访问中，v就已经出现过了，有环
                    return False
                if v not in visited:
                    #如果v还没有被访问过就递归地访问它
                    if not self.topologicalSort(adj,v,visited,loop,stack):
                        #如果发现了环，就返回
                        return False
        loop.remove(c)#这一轮内能访问的节点都访问完了，就从loop里边删除c
        stack.append(c)
        return True
````

## 772.基本计算器III
实现一个计算器来计算简单的表达式字符串：
表达式可以包含（）+-*/非负整数和空格，整数除法需要向下截断
给数范围 in [-2147483648,2147483647]

### 思路：
1. 加法器：  1+1=2  
优先队列放入字符一个一个遍历，是数字则放入num变量，是+则加入sum然后num清零，需要在优先队列最后添加一个+进行最后一步操作  
2. 空格：  
遇到即跳过  
3. 减法器： 1+1-10=-8
新增一个sign变量，表示符号，把-10变成+(-10)，初始化为+，遇-则变-，用sign变量判断是加是减
4. 乘除法：
需要考虑符号优先级了，利用堆栈stack，是加减就先放入堆栈，是乘除就取出堆栈里的数pop然后计算然后压入，最后堆栈内求和即可，减号负数着压栈
5. 括号：
使用递归，优先处理小括号里边的内容，左括号开始递归，右括号递归结束返回


````python
import collections
def calculate_expr(expr):
    q = collections.deque()
    for c in expr:
        if c!=' ':
            q.append(c)
    q.append('+')#这是为了最后一步操作
    def calculate(q):
        num = 0
        sign = '+'
        stack = []
        while q:
            c = q.popleft()
            if c.isdigit():
                num = 10*num+ord(c)-ord('0')#应对多位数
            elif c=='(':
                num = calculate(q)
            else:
                #已经到下一个符号了，该根据上一个sign来进行对上一个num操作了
                if sign=='+':
                    stack.append(num)
                elif sign=='-':
                    stack.append(-num)
                elif sign=='*':
                    stack.append(stack.pop()*num)
                elif sign=='/':
                    stack.append(stack.pop()/num)
                num = 0
                sign = c
                if c==')':
                    break
        return sum(stack)
    return calculate(q)
````

````python
#1.回溯法1——从前向后
class Solution:
    def isMatch(self, s: str, p: str) -> bool:
        if not p:
            return not s
        # 第一个字母是否匹配
        first_match = bool(s and p[0] in {s[0],'.'})
        # 如果 p 第二个字母是 *
        if len(p) >= 2 and p[1] == "*":
            return self.isMatch(s, p[2:]) or first_match and self.isMatch(s[1:], p)
        else:
            return first_match and self.isMatch(s[1:], p[1:])
        
class Solution:
    def isMatch(self, s: str, p: str) -> bool:
        if not s or not p:
            return False
        return self.match(s,0,p,0)#将指针i和j都指向0的位置
    
    def match(self,s,i,p,j):
        m = len(s)
        n = len(p)
        if j==n:
            return i==m
        if j==n-1 or p[j+1]!='*':
            return i<m and (p[j]=='.' or s[i]==p[j]) and self.match(s,i+1,p,j+1)
        if j<n-1 and p[j+1]=='*':
            while i<m and (p[j]=='.' or s[i]==p[j]):
                if self.match(s,i,p,j+2):
                    return True
                i+=1 #用星号去匹配更长的一段字符串
            if self.match(s,i,p,j+2):
                return True
            return False

                

#2.dp写法
#56 ms	13.7 MB
class Solution:
    def isMatch(self, s: str, p: str) -> bool:
        #竟然是dp
        lenp = len(p)
        lens = len(s)
        dp = [[False]*(lenp+1) for _ in range(lens+1)]
        dp[0][0]=True
        for j in range(1,lenp+1):
            if p[j-1]=='*':
                dp[0][j]=dp[0][j-1] or dp[0][j-2]#空串和非空正则，不能直接定义 true 和 false，必须要计算出来。
                #跟空字符串匹配要求都是一个星号在匹配
        for i in range(1,lens+1):
            for j in range(1,lenp+1):
                if s[i-1]==p[j-1] or p[j-1]=='.':
                    dp[i][j]=dp[i-1][j-1]#p和s当前字符匹配，进到下一字符
                elif p[j-1]=='*':
                    if s[i-1]==p[j-2] or p[j-2]=='.':#看再往前一个
                        #当前字符和前一个字符是匹配的
                        dp[i][j] = dp[i][j-2] or dp[i][j-1] or dp[i-1][j]#true/false传递
                        #1）删除前一个字符 dp[i][j-2]
                        #2）保留前一个字符 dp[i][j-1]
                        #3）复制前一个字符 dp[i-1][j]
                    else:
                        #尾上2个p的字符废了,只能删除字符
                        dp[i][j] = dp[i][j-2] 
        return dp[-1][-1]
````

## 222. 完全二叉树的节点个数
给出一个完全二叉树，求出该树的节点个数。

说明：

完全二叉树的定义如下：在完全二叉树中，除了最底层节点可能没填满外，其余每层节点数都达到最大值，并且最下面一层的节点都集中在该层最左边的若干位置。若最底层为第 h 层，则该层包含 1~ 2h 个节点。

示例:

输入: 
```python
    1
   / \
  2   3
 / \  /
4  5 6
```
输出: 6

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def countNodes(self, root: TreeNode) -> int:
        #要看左右子树的高度，完全二叉树的高度可以直接通过不断地访问左子树就可以获取
        #判断左右子树的高度: 
        #如果相等说明左子树是满二叉树, 然后进一步判断右子树的节点数(最后一层最后出现的节点必然在右子树中)
        #如果不等说明右子树是深度小于左子树的满二叉树, 然后进一步判断左子树的节点数(最后一层最后出现的节点必然在左子树中)
        if not root:
            return 0 
        left_depth = self.depth(root.left)
        right_depth = self.depth(root.right)
        if left_depth==right_depth:
            #左子树是满的，延伸到右子树去了，过了中间，接着检查右子树
            #节点数：2^depth-1 + 1 +rightnode
            return 2**left_depth + self.countNodes(root.right)
        else:
            #左子树可能没满，至少没能延伸到右边去，那么右子树是个少一层的满二叉树，检查左边就好
            return 2**right_depth + self.countNodes(root.left)

    def depth(self,node):
        depth = 0
        while node:
            depth+=1
            node = node.left #沿着左节点找下去
        return depth
````
