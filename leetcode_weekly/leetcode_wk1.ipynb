## 19. 删除链表的倒数第N个节点
给定一个链表，删除链表的倒数第 n 个节点，并且返回链表的头结点。

示例：
给定一个链表: 1->2->3->4->5, 和 n = 2.

当删除了倒数第二个节点后，链表变为 1->2->3->5.
说明：给定的 n 保证是有效的。

进阶：你能尝试使用一趟扫描实现吗？

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

#1.两次遍历法 O(L),O(1)
#36 ms	13.8 MB
#将添加一个哑结点作为辅助，该结点位于列表头部。哑结点用来简化某些极端情况，例如列表中只含有一个结点，或需要删除列表的头部。
class Solution:
    def removeNthFromEnd(self, head: ListNode, n: int) -> ListNode:
        p = head
        l = 0
        dummy = ListNode(0)
        dummy.next = head #把哑结点连接在head上
        while p:
            l += 1
            p = p.next
        l -= n
        p = dummy
        while l>0:
            l -= 1
            p = p.next
        p.next = p.next.next
        return dummy.next

#2.一次遍历法：快慢指针 O(L),O(1)
#40 ms	13.8 MB
#使用两个指针而不是一个指针。第一个指针从列表的开头向前移动n+1步，而第二个指针将从列表的开头出发。
#现在，这两个指针被n个结点分开，通过同时移动两个指针向前来保持这个恒定的间隔，直到第一个指针到达最后一个结点。
#此时第二个指针将指向从最后一个结点数起的第n个结点。重新链接第二个指针所引用的结点的 next 指针指向该结点的下下个结点。
class Solution:
    def removeNthFromEnd(self, head: ListNode, n: int) -> ListNode:
        dummy = ListNode(0)
        dummy.next = head
        p1 = dummy
        p2 = dummy
        for i in range(1,n+2):
            p1 = p1.next
            #这样中间刚好隔开两个
        while p1:
            p1 = p1.next
            p2 = p2.next
        p2.next = p2.next.next
        return dummy.next
#3. 奇妙递归法 O(L),?
#44 ms	13.8 MB
# 没有设立哑结点，但是有个计数的pos判断是否是头结点被去掉
class Solution:
    def removeNthFromEnd(self, head: ListNode, n: int) -> ListNode:
        def helper(node,n):
            if not node.next:
                return 0
            pos = helper(node.next,n)+1 #divide+conquer
            if pos == n:
                node.next = node.next.next
            return pos
        
        pos = helper(head,n)
        if pos == n-1:
            return head.next
        return head
    
```

## 23. 合并K个排序链表
合并 k 个排序链表，返回合并后的排序链表。请分析和描述算法的复杂度。

示例输入:
[
  1->4->5,
  1->3->4,
  2->6
]
输出: 1->1->2->3->4->4->5->6

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#1. PY3的优先队列with猴子补丁
#python3中如果类没有实现__lt__方法就不能进行比较，所以如果没有等号之间那段代码的话会报错
#TypeError: '<' not supported between instances of 'ListNode' and 'ListNode'，
#leetcode的ListNode类已经固定了，不能在上面直接加方法，于是打个猴子补丁搞定。

from queue import PriorityQueue
class Solution:
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        #=================================
        def com_list_node(a,b):
            return a.val < b.val
        ListNode.__lt__ = com_list_node
		#=================================
        head = p = ListNode(0)
        q = PriorityQueue()
        for l in lists:
            if l:
                q.put((l.val,l)) #l的head值和l的listnode_head
        while not q.empty():
            val,node = q.get()
            p.next = ListNode(val)
            p = p.next
            node = node.next
            if node:
                q.put((node.val, node))
        return head.next

#2. 快乐分治法 with 两两和一
class Solution(object):
    def mergeKLists(self, lists):
        """
        :type lists: List[ListNode]
        :rtype: ListNode
        """
        amount = len(lists)
        interval = 1
        while interval < amount:
            for i in range(0, amount - interval, interval * 2):
                lists[i] = self.merge2Lists(lists[i], lists[i + interval])
            interval *= 2 #其实是在说2分的次数,最后都合到第一个listnode上
        return lists[0] if amount > 0 else lists

    def merge2Lists(self, l1, l2):
        head = point = ListNode(0)
        while l1 and l2:
            if l1.val <= l2.val: #比较当前的元素大小
                point.next = l1 #小的放前边
                l1 = l1.next #下一个补上来
            else:
                point.next = l2 #小的放前边
                l2 = l2.next
            point = point.next #更新到下一个
        if not l1:
            point.next=l2
        else:
            point.next=l1
        return head.next
```

## 33. 搜索旋转排序数组
假设按照升序排序的数组在预先未知的某个点上进行了旋转。
( 例如，数组 [0,1,2,4,5,6,7] 可能变为 [4,5,6,7,0,1,2] )。
搜索一个给定的目标值，如果数组中存在这个目标值，则返回它的索引，否则返回 -1 。
你可以假设数组中不存在重复的元素。
**你的算法时间复杂度必须是 O(log n) 级别。**

示例 1:
输入: nums = [4,5,6,7,0,1,2], target = 0
输出: 4
示例 2:
输入: nums = [4,5,6,7,0,1,2], target = 3
输出: -1

```python
#1. 无脑法，但是不是O(logn)不满足要求
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        #list.in()方法的复杂度是O(n)
        return nums.index(target) if (target in nums) else -1
    
#2.正统递归法
#细心，看好边界等于号，看好各类判断条件
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        #试试递归
        def helper(nums,i,j,target):
            if i>j:
                return -1
            mid = (i+j)//2
            if nums[mid]==target:
                return mid
            if nums[i]>nums[mid]:
                #从mid到j有序，转折点在左半侧,可以借此判断tar的位置
                if nums[mid]<target and target<=nums[j]:
                    return helper(nums,mid+1,j,target)
                else:
                    return helper(nums,i,mid-1,target)
            else:
                #mid的左半侧有序，转折点在右侧,可以借此判断tar的位置
                if nums[i]<=target and target<nums[mid]:
                    return helper(nums,i,mid-1,target)
                else:
                    return helper(nums,mid+1,j,target)
        return helper(nums,0,len(nums)-1,target)

#3.二分查找法，就只是把递归用while替掉了而已
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        #二分搜索法
        n = len(nums)
        l = 0
        r = n-1
        while l<=r:
            mid = (l+r)//2
            if nums[mid]==target:
                return mid
            if nums[mid]<nums[r]:
                if nums[mid]<target and target<=nums[r]:
                    l = mid +1
                else:
                    r = mid-1
            else:
                if nums[l]<=target and target< nums[mid]:
                    r = mid -1
                else:
                    l = mid + 1
        return -1
```

## 50. Pow(x, n)
实现 pow(x, n) ，即计算 x 的 n 次幂函数。

示例 1:

输入: 2.00000, 10
输出: 1024.00000
示例 2:

输入: 2.10000, 3
输出: 9.26100
示例 3:

输入: 2.00000, -2
输出: 0.25000
解释: 2-2 = 1/22 = 1/4 = 0.25
说明:

-100.0 < x < 100.0
n 是 32 位有符号整数，其数值范围是 [−231, 231 − 1] 。


```python
#1.快速幂-循环
class Solution:
    def myPow(self, x: float, n: int) -> float:
        if n==0:
            return 1.0
        m = n
        if m<0:
            m=-m
        tmp = []
        while m>1:
            if (m & 1)==0:
                #偶数
                x *= x
                m //=2
            else:
                tmp.append(x)
                x *= x
                m = (m-1)>>1
        for i in range(len(tmp)):
            x *= tmp[i]
        if n<0:
            x = 1.0/x
        return x
#2.快速幂-递归
```

## 658. 找到 K 个最接近的元素
给定一个排序好的数组，两个整数 k 和 x，从数组中找到最靠近 x（两数之差最小）的 k 个数。返回的结果必须要是按升序排好的。如果有两个数与 x 的差值一样，优先选择数值较小的那个数。

示例 1:
输入: [1,2,3,4,5], k=4, x=3
输出: [1,2,3,4]

示例 2:
输入: [1,2,3,4,5], k=4, x=-1
输出: [1,2,3,4]
 

说明:
k 的值为正数，且总是小于给定排序数组的长度。
数组不为空，且长度不超过 10^4
数组里的每个元素与 x 的绝对值不超过 10^4

```python
#1.二分查找左边界法
class Solution:
    def findClosestElements(self, arr: List[int], k: int, x: int) -> List[int]:
        i = 0 
        j = len(arr)-k
        while i<j:
            mid = (i+j)//2
            if x-arr[mid]>arr[mid+k]-x:
                    i = mid+1
            else:
                    j = mid
            #区间定义左闭右开
        return arr[i:i+k]
```

## 34. 在排序数组中查找元素的第一个和最后一个位置
给定一个按照升序排列的整数数组 nums，和一个目标值 target。找出给定目标值在数组中的开始位置和结束位置。
你的算法时间复杂度必须是 O(log n) 级别。
如果数组中不存在目标值，返回 [-1, -1]。

示例 1:
输入: nums = [5,7,7,8,8,10], target = 8
输出: [3,4]
示例 2:
输入: nums = [5,7,7,8,8,10], target = 6
输出: [-1,-1]


```python
#1.两次二分查找，找左边界一次右边界一次
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        ans = [-1,-1]
        if len(nums)==0:
            return [-1,-1]
        i = 0
        j = len(nums)-1
        while i<j:
            mid = (i+j)//2
            if nums[mid]>=target:#在左边
                    j = mid
            else:
                    i = mid+1
        if nums[j]!=target:
            return ans
        ans[0]=i
        j = len(nums)
        while i<j:
            mid = (i+j)//2
            if nums[mid]<=target:#在右边
                    i = mid+1
            else:
                    j = mid
        ans[1]=j-1
        return ans
```

## 153. 寻找旋转排序数组中的最小值
假设按照升序排序的数组在预先未知的某个点上进行了旋转。

( 例如，数组 [0,1,2,4,5,6,7] 可能变为 [4,5,6,7,0,1,2] )。

请找出其中最小的元素。
你可以假设数组中不存在重复元素。

示例 1:
输入: [3,4,5,1,2]
输出: 1
示例 2:
输入: [4,5,6,7,0,1,2]
输出: 0

```python
#1.寻找转折点
class Solution:
    def findMin(self, nums: List[int]) -> int:
        #其实是寻找谷值呢
        i = 0
        j = len(nums)-1
        while i<j:
            mid = (i+j)//2
            if nums[mid]<nums[j]:#mid右侧有序，转折点在左边
                j= mid
            else:#mid左侧有序，转折点在mid和j之间
                i = mid+1
        return nums[j]
```

## 162. 寻找峰值
峰值元素是指其值大于左右相邻值的元素。

给定一个输入数组 nums，其中 nums[i] ≠ nums[i+1]，找到峰值元素并返回其索引。

数组可能包含多个峰值，在这种情况下，返回任何一个峰值所在位置即可。

你可以假设 nums[-1] = nums[n] = -∞。

示例 1:

输入: nums = [1,2,3,1]
输出: 2
解释: 3 是峰值元素，你的函数应该返回其索引 2。
示例 2:

输入: nums = [1,2,1,3,5,6,4]
输出: 1 或 5 
解释: 你的函数可以返回索引 1，其峰值元素为 2；
     或者返回索引 5， 其峰值元素为 6。


```python
#左闭右开
class Solution:
    def findPeakElement(self, nums: List[int]) -> int:
        i = 0
        j = len(nums)-1
        while i<j:
            mid=(i+j)//2
            if nums[mid]>nums[mid+1]:#mid可能是
                j=mid
            else:#mid不会是
                i = mid + 1
        return i
```

## 1095. 山脉数组中查找目标值
（这是一个 交互式问题 ）

给你一个 山脉数组 mountainArr，请你返回能够使得 mountainArr.get(index) 等于 target 最小 的下标 index 值。
如果不存在这样的下标 index，就请返回 -1。
何为山脉数组？如果数组 A 是一个山脉数组的话，那它满足如下条件：
首先，A.length >= 3
其次，在 0 < i < A.length - 1 条件下，存在 i 使得：

A[0] < A[1] < ... A[i-1] < A[i]
A[i] > A[i+1] > ... > A[A.length - 1]
 
你将 不能直接访问该山脉数组，必须通过 MountainArray 接口来获取数据：
MountainArray.get(k) - 会返回数组中索引为k 的元素（下标从 0 开始）
MountainArray.length() - 会返回该数组的长度
 
注意：

对 MountainArray.get 发起超过 100 次调用的提交将被视为错误答案。此外，任何试图规避判题系统的解决方案都将会导致比赛资格被取消。
为了帮助大家更好地理解交互式问题，我们准备了一个样例 “答案”：https://leetcode-cn.com/playground/RKhe3ave，请注意这 不是一个正确答案。


示例 1：
输入：array = [1,2,3,4,5,3,1], target = 3

示例 2：
输入：array = [0,1,2,4,2,1], target = 3
输出：-1
解释：3 在数组中没有出现，返回 -1。
 

提示：
3 <= mountain_arr.length() <= 10000
0 <= target <= 10^9
0 <= mountain_arr.get(index) <= 10^9

```python
#1.堆叠二分查找--峰值问题arrange
# """
# This is MountainArray's API interface.
# You should not implement it, or speculate about its implementation
# """
#class MountainArray(object):
#    def get(self, index):
#        """
#        :type index: int
#        :rtype int
#        """
#
#    def length(self):
#        """
#        :rtype int
#        """

class Solution(object):
    def findInMountainArray(self, target, mountain_arr):
        """
        :type target: integer
        :type mountain_arr: MountainArray
        :rtype: integer
        """
        #try二分搜索
        i=0
        j=mountain_arr.length()-1
        while i<j:
            mid = (i+j)//2
            arr_midd = mountain_arr.get(mid)
            arr_midd_1 = mountain_arr.get(mid+1)
            if arr_midd>arr_midd_1:
                j = mid
            else:
                i = mid+1
        sep_pnt = i
        i = 0
        j = sep_pnt
        while i<=j:
            mid = (i+j)//2
            arr_midd = mountain_arr.get(mid)
            if arr_midd==target:
                return mid
            if arr_midd<target:
                i = mid+1
            else:
                j = mid-1
        i = sep_pnt+1
        j = mountain_arr.length()-1
        while i<=j:
            mid = (i+j)//2
            arr_midd = mountain_arr.get(mid)
            if arr_midd==target:
                return mid
            if arr_midd>target:
                i = mid+1
            else:
                j = mid-1
        return -1
```

## 121. 买卖股票的最佳时机
给定一个数组，它的第 i 个元素是一支给定股票第 i 天的价格。
如果你最多只允许完成一笔交易（即买入和卖出一支股票一次），设计一个算法来计算你所能获取的最大利润。
注意：你不能在买入股票前卖出股票。

示例 1:

输入: [7,1,5,3,6,4]
输出: 5
解释: 在第 2 天（股票价格 = 1）的时候买入，在第 5 天（股票价格 = 6）的时候卖出，最大利润 = 6-1 = 5 。
     注意利润不能是 7-1 = 6, 因为卖出价格需要大于买入价格；同时，你不能在买入前卖出股票。
示例 2:

输入: [7,6,4,3,1]
输出: 0
解释: 在这种情况下, 没有交易完成, 所以最大利润为 0。

```python
#dp
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        if len(prices)<=1:
            return 0
        dp = 0
        for i in range(1,len(prices)):
            dp = max(dp,prices[i]-min(prices[:i]))
        return dp
```

```python
#1.dp-贪心同理
import sys
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        mm=-sys.maxsize
        dp = [0 for _ in range(len(nums)+1)]
        for i in range(1,len(nums)-1):
            dp[i] = max(dp[i-1]+nums[i-1],nums[i-1]) #决定是否重新开始,两个i意味不同
            if dp[i]>mm:
                mm = dp[i]
        return mm

#2.分治法
class Solution:
    def cross_sum(self, nums, left, right, p): 
            if left == right:
                return nums[left]
            left_subsum = float('-inf')
            curr_sum = 0
            for i in range(p, left - 1, -1):
                curr_sum += nums[i]
                left_subsum = max(left_subsum, curr_sum)

            right_subsum = float('-inf')
            curr_sum = 0
            for i in range(p + 1, right + 1):
                curr_sum += nums[i]
                right_subsum = max(right_subsum, curr_sum)

            return left_subsum + right_subsum   
    
    def helper(self, nums, left, right): 
        if left == right:
            return nums[left]
        
        p = (left + right) // 2
            
        left_sum = self.helper(nums, left, p)
        right_sum = self.helper(nums, p + 1, right)
        cross_sum = self.cross_sum(nums, left, right, p)
        
        return max(left_sum, right_sum, cross_sum)
        
    def maxSubArray(self, nums: 'List[int]') -> 'int':
        return self.helper(nums, 0, len(nums) - 1)
```

```python
#1.list-if查看法：一次遍历
#84 ms	13.7 MB
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        #感觉得完全遍历一遍
        s = list(s)
        if len(s)==0:
            return 0
        tmp=[]
        m_len = 0
        for i in range(len(s)):
            if s[i] not in tmp:
                tmp.append(s[i])
                m_len = len(tmp) if len(tmp)>m_len else m_len
            else:
                m_len = len(tmp) if len(tmp)>m_len else m_len
                tmp= tmp[tmp.index(s[i])+1:] + [s[i]]
        return m_len

#2.set哈希：for-while法
#92 ms	13.7 MB
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        # 哈希集合，记录每个字符是否出现过
        occ = set()
        n = len(s)
        # 右指针，初始值为 -1，相当于我们在字符串的左边界的左侧，还没有开始移动
        rk, ans = -1, 0
        for i in range(n):
            if i != 0:
                # 左指针向右移动一格，移除一个字符
                occ.remove(s[i - 1])
            while rk + 1 < n and s[rk + 1] not in occ:
                # 不断地移动右指针
                occ.add(s[rk + 1])
                rk += 1
            # 第 i 到 rk 个字符是一个极长的无重复字符子串
            ans = max(ans, rk - i + 1)
        return ans

#3.字典哈希：一次遍历法
#52 ms	13.7 MB
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        k, res, c_dict = -1, 0, {}
        for i, c in enumerate(s):
            if c in c_dict and c_dict[c] > k:  # 字符c在字典中 且 上次出现的下标大于当前长度的起始下标，在当前的最长字符串儿里边
                k = c_dict[c]
                c_dict[c] = i
            else:
                c_dict[c] = i
                res = max(res, i-k)
        return res
```

为了表示给定链表中的环，我们使用整数 pos 来表示链表尾连接到链表中的位置（索引从 0 开始）。 如果 pos 是 -1，则在该链表中没有环。
示例 1：

输入：head = [3,2,0,-4], pos = 1
输出：true
解释：链表中有一个环，其尾部连接到第二个节点。
示例 2：
输入：head = [1,2], pos = 0
输出：true
解释：链表中有一个环，其尾部连接到第一个节点。


```python
#1.快慢指针：追逐与重合
class Solution:
    def hasCycle(self, head: ListNode) -> bool:
        #难道是快慢指针
        #p1 = ListNode(0)
        p1 = head
        p2 = head
        while p1 and p1.next:
            p1 = p1.next.next
            p2 = p2.next
            if p1==p2:
                return True
        return False
#2.哈希法：遍历一圈儿，用完就删，及时停止
class Solution:
    def hasCycle(self, head: ListNode) -> bool:
        if not head:
            return False
        while head.val!=None and head.next:
            head.val = None
            head = head.next
        if not head.next:
            return False
        return True
```

示例:

输入: [2,3,1,1,4]
输出: 2
解释: 跳到最后一个位置的最小跳跃数是 2。
     从下标为 0 跳到下标为 1 的位置，跳 1 步，然后跳 3 步到达数组的最后一个位置。
说明:

假设你总是可以到达数组的最后一个位置。


```python
#1.DP
#64 ms	15.5 MB
class Solution:
    def jump(self, nums: List[int]) -> int:
        if len(nums)<2:
            return 0
        #感觉是dp
        i = 0
        j = 1
        maxjump = 0
        dp = [0]*len(nums)
        while j<len(nums):
            maxjump = i+nums[i]
            while j<len(nums) and maxjump>=j:
                dp[j] = dp[i]+1
                j+=1
            i+=1
        return dp[-1]
#2.贪心：正向更新可达到边界
#48 ms	15.2 MB
class Solution:
    def jump(self, nums: List[int]) -> int:
        n = len(nums)
        maxPos, end, step = 0, 0, 0
        for i in range(n - 1):
            if maxPos >= i:
                maxPos = max(maxPos, i + nums[i])
                if i == end:
                    end = maxPos
                    step += 1
        return step
```
